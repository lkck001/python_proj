```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Define the vector function T(u,v)
def T(u, v):
    return (u * np.cos(v), u * np.sin(v), u**2 + v**2)

# 2. Compute the partial derivatives with respect to u and v
def Tu(u, v):
    return (np.cos(v), np.sin(v), 2*u)

def Tv(u, v):
    return (-u * np.sin(v), u * np.cos(v), 2*v)

# 3. Compute the normal vector to the plane at (1,0)
u_val = 1
v_val = 0
normal_vector = np.cross(Tu(u_val, v_val), Tv(u_val, v_val))

# 4. Plot T(u,v)
u = np.linspace(-2, 2, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

x, y, z = T(u, v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.7)

# 5. Plot the plane at (1,0)
point_on_plane = T(u_val, v_val)
d = -point_on_plane[0] * normal_vector[0] - point_on_plane[1] * normal_vector[1] - point_on_plane[2] * normal_vector[2]
xx, yy = np.meshgrid(np.linspace(-2, 2, 100), np.linspace(-2, 2, 100))
zz = (-normal_vector[0] * xx - normal_vector[1] * yy - d) * 1. / normal_vector[2]
ax.plot_surface(xx, yy, zz, color='green', alpha=0.5)

# Add the planePoint to the plot
ax.scatter(*point_on_plane, color='red', s=100)

# Show the plot
plt.show()
```
